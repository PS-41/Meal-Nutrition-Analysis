```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_path = "../data"
label_train_path = f"{base_path}/label_train.csv"

label_train = pd.read_csv(label_train_path)

print(label_train.head())
print(label_train.info())
```

```python
# Check for missing values in the label_train dataset
missing_values = label_train.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
# Plot distribution of Lunch Calories and Breakfast Calories
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(label_train['Lunch Calories'], bins=30, kde=True)
plt.title("Distribution of Lunch Calories")

plt.subplot(1, 2, 2)
sns.histplot(label_train['Breakfast Calories'], bins=30, kde=True)
plt.title("Distribution of Breakfast Calories")

plt.show()
```

```python
# Compute and visualize the correlation matrix
correlation_matrix = label_train.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix for Nutritional Data")
plt.show()
```

```python
# Scatter plot of Breakfast Calories vs Lunch Calories
sns.scatterplot(x='Breakfast Calories', y='Lunch Calories', data=label_train)
plt.title("Breakfast Calories vs Lunch Calories")
plt.xlabel("Breakfast Calories")
plt.ylabel("Lunch Calories")
plt.show()
```

```python
# Boxplot for Lunch Calories
plt.figure(figsize=(8, 5))
sns.boxplot(y=label_train['Lunch Calories'])
plt.title("Boxplot of Lunch Calories")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from ast import literal_eval

# Load the CGM data from CSV
cgm_train = pd.read_csv("../data/cgm_train.csv")

# Select a single subject and day for visualization
subject_id = img_train['Subject ID'][0]  # Example: first subject in the dataset
day = img_train['Day'][0]  # Example: first day for this subject

# Filter the CGM data for the chosen subject and day
subject_data = cgm_train[(cgm_train['Subject ID'] == subject_id) & (cgm_train['Day'] == day)]

# Convert the string representation of the 'CGM Data' to an actual list of tuples
cgm_data_str = subject_data['CGM Data'].iloc[0]
cgm_data = literal_eval(cgm_data_str)  # Convert string to list of tuples

# Separate times and glucose values
times = [pd.to_datetime(time) for time, glucose in cgm_data]
glucose_values = [float(glucose) for time, glucose in cgm_data]

# Plot glucose values over time
plt.figure(figsize=(10, 5))
plt.plot(times, glucose_values, label='Glucose Level')
plt.xlabel('Time')
plt.ylabel('Glucose Level')
plt.title(f'Glucose Levels for Subject {subject_id} on Day {day}')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from ast import literal_eval

# Load the demographic and microbiome data from CSV
demo_viome_train = pd.read_csv("../data/demo_viome_train.csv")

# Convert 'Viome' column from string to list of floats
demo_viome_train['Viome'] = demo_viome_train['Viome'].apply(lambda x: literal_eval(x) if isinstance(x, str) else x)

# Expand the 'Viome' column into separate columns
viome_expanded = pd.DataFrame(demo_viome_train['Viome'].tolist(), index=demo_viome_train.index)
viome_expanded.columns = [f'Viome_{i+1}' for i in range(viome_expanded.shape[1])]  # Name columns as Viome_1, Viome_2, etc.

# Concatenate the expanded columns with the original dataframe (excluding the old 'Viome' column)
demo_viome_train_expanded = pd.concat([demo_viome_train.drop(columns=['Viome']), viome_expanded], axis=1)

# Display the first few rows to verify
demo_viome_train_expanded.head()
```

```python
# Check for missing values in each column
missing_values = demo_viome_train_expanded.isnull().sum()

# Display columns with missing values, if any
missing_values = missing_values[missing_values > 0]
print("Columns with missing values:")
print(missing_values)
```
